```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Agriculture Data Engineering Pipeline") \
    .getOrCreate()

print("Spark Session Created ✅")
```

```python
data = [
    ("India", "Rice", 2023, 30, 3.5),
    ("USA", "Wheat", 2023, 25, 4.0),
    ("India", "Maize", 2023, None, 2.8),
    ("USA", "Corn", 2023, 28, None)
]

columns = ["Country", "Crop", "Year", "Temperature", "Yield"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
from pyspark.sql.functions import col


clean_df = df.dropna()

filtered_df = clean_df.filter(col("Year") == 2023)

filtered_df.show()
```

```python

filtered_df.write.mode("overwrite").parquet("/content/agriculture_data_lake")

print("Data stored in Data Lake ✅")
```

```python

lake_df = spark.read.parquet("/content/agriculture_data_lake")

lake_df.show()
```
